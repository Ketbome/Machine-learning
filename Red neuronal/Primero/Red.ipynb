```python
# TensorFlow / Pytorch
import pandas as pd

datos = pd.read_csv('celsius.csv')
```

```python
datos.head()
```

```python
datos["celsius"].values
```

```python
X = datos["celsius"].values
y = datos["fahrenheit"].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)), # 1 entrada, los grados celsius: 1 valor
    tf.keras.layers.Dense(units=1), # 1 neurona, la salida: 1 valor    
])

modelo.compile(
    optimizer=tf.keras.optimizers.Adam(.1),
    loss='mean_squared_error'
)
```

```python
entrenamiento = modelo.fit(X, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.plot(entrenamiento.history["loss"])
```

```python
c = 134
prediccion = modelo.predict([c])
print(f"Predicción: {prediccion} Fahrenheit para {c} Celsius") 
```
